```python
# COLUMN BY COLUMN CLEANING

# FOUNDED COLUMN CLEANING


# CHECKING FOR MISSING VALUES
startup_2021['Founded'].isna().sum()
```

### the Founded Column has only 1 nan value. We can use the mean or the mode to fill it

```python
startup_2021['Founded'].unique()
```

```python
# SETTING THE DATATYPE FOR THE FOUNDED COLUMN TO STRING

startup_2021['Founded'] = startup_2021['Founded'].replace(np.nan, 0)

startup_2021['Founded'] = startup_2021['Founded'].astype(int)

startup_2021['Founded']
```

```python
# We are filling the "NAN" with the mean 

# calculating the mean and repalcing "0" with it

mean_val = startup_2021['Founded'].mean()

mean_val
```

```python
startup_2021.Founded = startup_2021.Founded.replace("0",2015)

startup_2021.Founded
```

```python
startup_2021.Founded.isna().sum()
```

### Our Founded Column is clean now.

```python
# CLEANING OF THE HEADQUARTERS COLUMN

startup_2021['HeadQuarter'].unique()
```

```python
# checking the missing value

startup_2021['HeadQuarter'].isna().sum()
```

### It has only one missing value

```python
# Replacing NAN values with "Undefined"

headquarters_2021 = startup_2021.HeadQuarter.isnull()

headquarters_2021
```

```python
# Replacing the 'NAN' value with Keyword "Undefined"

for i in range(len(headquarters_2021)):
    if headquarters_2021[i] == True:
        startup_2021.HeadQuarter[i] = "Undefined"
```

```python
startup_2021.HeadQuarter.isnull().sum()
```

```python
startup_2021['HeadQuarter'] = startup_2021['HeadQuarter'].str.replace('Pharmaceuticals\t#REF!', 'Pharmaceuticals')

startup_2021['HeadQuarter'] = startup_2021['HeadQuarter'].str.replace('Gurugram\t#REF!', 'Gurugram')

startup_2021['HeadQuarter'] = startup_2021['HeadQuarter'].str.replace('Online Media\t#REF!', 'Online Media')


startup_2021
```

```python
startup_2021.isna().sum()
```

### Now, the Headquarters Column is great.
